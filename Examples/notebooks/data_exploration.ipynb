```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv(Path('C:/Projects/University/Semester 2 Projects/Energy-efficient Distributed Systems'
                   '/energy_efficient_ds/Assignment P1/data/web_traffic_wikipedia_tiny.csv'))
# Rearrange/fix dataset
df = df.transpose()
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
df.index.name = 'date'

# new style method 2; use an axes array
fig, axs = plt.subplots(2, 2)
fig.tight_layout()

df['Google_zh.wikipedia.org_all-access_all-agents']\
    .plot(ax = axs[0, 0], title = 'Google_zh.wikipedia.org')
df['Albania_en.wikipedia.org_all-access_all-agents']\
    .plot(ax = axs[0, 1], title = 'Albania_en.wikipedia.org')
df['Kosovo_en.wikipedia.org_all-access_all-agents']\
    .plot(ax = axs[1, 0], title = 'Kosovo_en.wikipedia.org')
df['Cesare_Maldini_fr.wikipedia.org_all-access_all-agents']\
    .plot(ax = axs[1, 1], title = 'Cesare_Maldini_fr.wikipedia.org')

plt.savefig('web_traffic.png', dpi = 300)
```
